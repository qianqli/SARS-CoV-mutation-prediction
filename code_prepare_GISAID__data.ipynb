## this code is to generate the acession number from GISAID with the format 'EPI-ISL_....'.  As for general users, you are only allowed to downloads 5000 submission once.  Here I checked the first submission is EPI_ISL_402119 on 10th, Jan, 2020. And the acession number is a series data based on this one (+1). 

```python
# code to generate a set of data. 
# there is error for 5000 range, using 4000 get all the data
accession_ID = "EPI_ISL_402119"
pre = "EPI_ISL_"
original_number = 402120
for i in range(4000):
  acc_num = pre + str(original_number + i)
  accession_ID += "," + acc_num
 
print(accession_ID)
```
